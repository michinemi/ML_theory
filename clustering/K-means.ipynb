```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
# Load the data
iris = load_iris()
X = iris.data[:, :4]
```

<h4>• n_clusters: number of clusters to form (in our case 3)</h4>
<h4>• max_iter: maximum number of iterations for the algorithm to converge (in our case 10)</h4>
<h4>• n_init: specifies the number of times the K-means algorithm will be run with different centroid seeds (in our case 10)</h4>

```python
# Fit the K-means model
kmeans = KMeans(n_clusters=3, max_iter=10, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Visualize the clusters
plt.figure(figsize=(5, 3))
colors = ['r', 'g', 'b']
for i in range(3): # n_clusters=3
    plt.scatter(X[labels == i, 0], X[labels == i, 1], c=colors[i], label=f'Cluster {i + 1}')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, c='k', label='Centroids')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('K-means Clustering of Iris Dataset')
plt.legend()
plt.show()
```
