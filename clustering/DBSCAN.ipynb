```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
```

```python
# Load the data
iris = datasets.load_iris()
X = iris.data[:, :4]
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

<h4>• eps: The maximum distance between two samples for them to be considered as in the same neighborhood (in our case 0.5)</h4>
<h4>• min_samples: The number of samples in a neighborhood for a point to be considered as a core point (in our case 5)</h4>

```python
# Implement DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)
```

```python
# Visualize the clusters with legend
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')

colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']
for cluster_label in np.unique(clusters):
    cluster_mask = (clusters == cluster_label)
    ax.scatter(X_scaled[cluster_mask, 0], X_scaled[cluster_mask, 1], X_scaled[cluster_mask, 2],
               c=colors[cluster_label % len(colors)], label=f'Cluster {cluster_label}', s=50)

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')

plt.title('DBSCAN Clustering of Iris Dataset')
plt.legend()
plt.show()
```
