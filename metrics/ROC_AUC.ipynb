```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import roc_auc_score
```

```python
file_path = '../data/random_data.csv'

df = pd.read_csv(file_path)
df = df.drop('y_pred_random_forest', axis=1)
df.rename(columns={'y_pred_logistic': 'y_pred'}, inplace=True)
df.head()
```

```python
fpr, tpr, thresholds = roc_curve(df.y_act, df.y_pred)
roc_auc = auc(fpr, tpr)
```

```python
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend()
plt.show()
```

```python
# Calculate the AUC score
auc_score = roc_auc_score(df.y_act, df.y_pred)
print("AUC score:", auc_score)
```
